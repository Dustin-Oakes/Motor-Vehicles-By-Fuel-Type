```python
    # import modules
import matplotlib.pyplot as plt
import pandas as pd
    
    #set up Data
Data = pd.read_csv("../DataSource/clean_car_data.csv")

Data.head()
```

```python
    # Sort Data by Make, and Brands who sold > 1000 units  
Make=Data['Make'].value_counts()
clean_data = Make.loc[(Make>1000)]

clean_data.head
```

```python
    # Map out Bar chart
fig, ax = plt.subplots()

    # Set x and y Axis
x = clean_data.index
y = clean_data.values

    #Color schyme
ax.barh(x,y, color='coral')
ax.set_facecolor("skyblue")

    # Bar Edits
ax.invert_yaxis()
ax.set_title('Most Popular Eletric/hybrid Brand in Washington')
ax.set_xlabel('Units Sold')
plt.yticks(rotation=50)

plt.savefig('../David/Bar_Chart1.png')
```
