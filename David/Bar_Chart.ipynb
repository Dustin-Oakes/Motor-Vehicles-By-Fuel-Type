```python
import matplotlib.pyplot as plt
import pandas as pd

Data = pd.read_csv("../David/Make_by_Washington.csv")
Data.head()
```

```python
    # Set x and y Axis
x=Data['Make']
y=Data['Sales']

    # Map out Bar chart
fig, ax = plt.subplots(figsize=(16, 9))
ax.barh(x,y,color='lightGreen')

    # Bar Edits
ax.invert_yaxis()
ax.set_title('Most Popular Brand in Washington', fontsize=20)
ax.set_xlabel('Sales', fontsize=15)

fig.text(0.75, 0.13, 'Brands selling over 1,000 units')

plt.savefig('../David/Bar_Chart1.png')
```
